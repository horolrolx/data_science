```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell': 'U.S.',
    'Apple': 'U.S.',
    'Acer': 'Taiwan',
    'HP': 'U.S.',
    'Lenovo': 'China',
    'Alienware': 'U.S.',
    'Microsoft': 'U.S.',
    'Asus': 'Taiwan'}
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.describe()
```

```python
nation_groups.size()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```

```python
import pandas as pd
df = pd.read_csv('data/occupations.csv')
```

```python
df.dtypes
```

```python
df.loc[df['gender'] == 'M', 'gender'] = 0
df.loc[df['gender'] == 'F', 'gender'] = 1
df['gender'] = df['gender'].astype(int)
occupation_group = df.groupby('occupation')['gender']
occupation_group.mean().sort_values(ascending=False)
```
